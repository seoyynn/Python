#### list 리스트
- 파이썬 자료구조 중 하나
- 순서가 있고, 수정 가능한 자료구조
- 대괄호([ ])로 작성되고, 리스트 내부의 값은 콤마(,)로 구분
- 하나의 리스트에 다양한 자료형을 포함
- 추가, 수정, 삭제가 가능하다

##### 리스트 선언

```python
# 리스트명 = [요소1, 요소2, 요소3, ...]
a = [] # 빈 리스트
b = [1, 2, 3] # 숫자요소를 담은 리스트
c = ['장서연', '장두식,' '조연경바보'] # 문자열 요소를 담은 리스트
d = [1, 2, '동동', '구리구리', '냥냥'] # 숫자, 문자열 요소를 담은 리스트
e = [1, 2, '동동', ['구리구리', '냥냥']] # 리스트 내부에 리스트를 포함한 리스트 

# [[]] : 이중리스트
# [] : 단일리스트
```

```python
print (a)
print (b)
print (c)
print (d)
print (e)
```

##### 리스트 데이터 접근 : 인덱싱, 슬라이싱

```python
# 인덱싱 : 리스트명[인덱스]
list1 = [1, 3, 5, 7, 9]
# 인덱스 번호 0번에 접근
list1[0]
# 7요소에 접근
print(list1[3])
```

```python
list2 = [1, 2, 3, ['a', 'b', 'c']]
temp = list2[3]

print(temp[1])
print(list2[3][1])
```

- 슬라이싱

```python
# 리스트명[시작인덱스 : 끝인덱스]
# 시작인덱스 <= 인덱스 번호 < 끝인덱스
list3 = ['장서연', '장두식,', '조연경바보', '박서연바보', '한가연']
# 조연경바보, 박서연도바보 데이터 접근 출력
list3[2:4] # 출력결과도 리스트로 반환
```

```python
# '장서연' ~ '조연경바보' 접근해보기
list3[0:3]
```

```python
# '장두식' ~ '박서연바보'
list3[1:]
```

```python
# 인덱스 번호 0부터 2개씩 띄어서 가져오기
list3[::2]
list3[1::3] # 인덱스 1번부터 끝까지 3개씩 띄어서 가져오기
```

```python
# 이중리스트를 만들고 데이터 접근해보기
# 이름, 나이, 혈액형 정보기반의 이중 리스트

list4 = [['장서연',27,'B'],
         ['조연경',27,'A'],
         ['박서연',21,'B'],
         ['박유정',28,'지각형'],
         ['김종현',32,'형']]

list4
```

```python
print(f'박서연씨의 나이는 {list4[2][1]}이고, 혈액형은 {list4[2][2]}형이에용')
```

```python
# 조연경이의 정보와 박서연이의 정보를 리스트로 가져오기
# 나이와 혈액형을 출력해보기
list4[1:3]
```

#### 튜플 tuple
- 파이썬의 자료구조 형태 중 하나
- 순서가 있는 집합 (인덱싱, 슬라이싱 가능)
- 소괄호 ()로 작성되어지고, 튜플 내부의 값은 콤마 ','로 구분
- 하나의 튜플에는 다양한 자료형을 포함
- 데이터의 추가, 수정, 삭제가 불가능하다.

##### 튜플 선언

```python
a = () # 빈 튜플
b = (1, 2, 3) # 숫자형 데이터를 담은 튜플
c = ('장서연','조연경','박서연') # 문자열 데이터를 담은 튜플
d = (1, 2, '박유정','김종현') # 숫자, 문자열 데이터를 담은 튜플
e = (('인공지능', '윤효창', 30),('A반', '오경진', 33)) # 이중튜플

print(a)
print(b)
print(c)
print(d)
print(e)
```

##### 튜플 데이터 접근 : 인덱싱, 슬라이싱

```python
# 튜플명[인덱스번호]
# 윤효창씨 나이 접근
e[0][2]
```

```python
# 이름과 나이 가져오기(슬라이싱)
e[0][1:]
```
