#### if문

```python
if True:
    print('실행문장 실행')
```

```python
# if문에서 조건식이 False였을때는 아무것도 실행하지 않고 끝이 난다
if False:
    print('실행문장 실행')
```

##### 들여쓰기의 중요성

```python
if True:
    print('실행문장을 실행')
print('if문 밖에있는 실행문장')
```

```python
# if문 밖에있는 들여쓰기를 벗어난 실행문장만 False 영향을 받지안흔ㄴ다
if False:
    print('실행')
print('실실행')
```

#### if ~ else 문
- 들여쓰기 주의!

```python
# 조건문 if-else를 활용해서 money가 10000원 이상이면 '택시' 10000원 이하이면 '버스'를 탄다를 출력해보기
money = 9000

if money >= 10000:
    print('택시')
else:
    print('버스')
```

```python
num = int(input('정수 입력 >> '))

if num%3==0 and num%5==0:
    print('3과 5의 배수입니다')
else:
    print('3과 5의 배수가 아닙니다')
```

```python
mask = int(input('마스크 개수 입력 >> '))
a = mask//8
b = mask%8//5+1


if a >= 1:
    print(f'8개들이 포장지 개수 : {a}')
    print(f'5개들이 포장지 개수 : {b}')
else:
    print(f'8개들이 포장지 개수 : {a}')
    print(f'5개들이 포장지 개수 : {b}')
```

```python
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))

if num1 > num2:
    print('첫 번째 정수가 더 큽니다')
elif num1 == num2:
    print('두 수가 똑같습니다')
else:
    print('두 번째 정수가 더 큽니다')
```

```python
score = int(input('점수 입력 >> '))

if score >= 90:
    print(f'{score}점은 A학점 입니다')
elif score >= 80:
    print(f'{score}점은 B학점 입니다')
elif score >= 70:
    print(f'{score}점은 C학점 입니다')
elif score >= 60:
    print(f'{score}점은 D학점 입니다')
else:
    print(f'{score}점은 F학점 입니다')
```

```python
print("""
================================================
================---- 자판기 ----================
================================================
== 1.콜라 == 2.사이다 ==  3.물  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00원===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++     SMHRD    +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===""")

money = int(input('금액 넣기 >> '))
select = int(input('메뉴 선택(번호 입력) >> '))

if money <= 500:
    print('금액이 부족합니다')
elif select == 1:
    print(f'1000원 : {(money-600)//1000}개 500원 : {(money-600)%1000//500}개, 100원 : {(money-600)%1000%500//100}개')
elif select == 2:
    print(f'1000원 : {(money-800)//1000}개 500원 : {(money-800)%1000//500}개, 100원 : {(money-800)%1000%500//100}개')
elif select == 3:
    print(f'1000원 : {(money-1000)//1000}개 500원 : {(money-1000)%1000//500}개, 100원 : {(money-1000)%1000%500//100}개')
else:
    print('다시 입력해주세용')
```

```python
print("""
================================================
================---- 자판기 ----================
================================================
== 1.콜라 == 2.사이다 ==  3.물  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00원===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++     SMHRD    +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===""")

money = int(input('금액 넣기 >> '))
select = int(input('메뉴 선택(번호 입력) >> '))
change = 0

if select == 1:
    change = money - 600
if select == 2:
    change = money - 800
if select == 3:
    change = money - 1000
    
ten, five, one = 0, 0, 0

if change > 0:
    ten = change // 1000
    five = change % 1000 // 500
    one = change % 1000 % 500 // 100
    print(f'1000원 : {ten}개, 500원 : {five}개, 100원 : {one}개')
else:
    print('금액이 부족합니다')
```
