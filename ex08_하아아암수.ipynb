#### 함수 기본 구조 4가지
- 매개변수 x, return x
- 매개변수 O, return x
- 매개변수 x, return O
- 매개변수 O, return O

```python
# 1. 매개변수 X, return X
def hi():
    print('앙용')

# 함수 호출
hi()
result = hi() # return 값이 없으면 null이 반환됨
print(result) # return 값이 없으면 대입을 하지 않음, 출력값을 확인하기 위해 print()
```

```python
# 3. 매개변수 X, return O
def hi3():
    return '안용'

# 함수 호출
hi3()
result = hi3()
print(result)
```

```python
# 4. 매개변수 O, return O
def hi4(name):
    return f'{name} 안뇽이야'

# 함수 호출
hi4('두시기야')
result = hi4('두시기야')
print(result)
```

```python
# return을 사용하여 함수를 정의할 때
def divisor(num):
    for i in range(1, num+1):
        if num % i == 0:
            return i
        
# return값을 사용하면 함수를 수행하면서 값 하나만 반환하고 종료
```

```python
divisor(10)
```

```python
# return을 사용하여 함수를 정의하고 싶을 때
# ㄴ> 나중에 결과값을 사용하기 위해 별도의 변수에 저장
def divisor(num):
    list1 = [] # 비어있는 리스트 선언
    for i in range(1, num+1):
        if num % i == 0:
            list1.append(i) # list안에 조건에 맞는 i값을 append사용해서 집어넣기
    return list1
    
divisor(10)
            
```

#### return 키워드 정리
- 반환값은 언제나 하나
    ㄴ> ex) num_add, num_sub = number_m(num1, num2)
        --> 결과값은 (num_add, num_sub)
    --> 매개변수 2개를 받아서 하나는 add 하나는 sub에 저장해주고 결과값은 패킹된 형태로 반환해준다

- 함수 안에서 return은 1번 실행
- 함수 안에서 return을 만나면 함수가 끝난다

```python
# 더하고, 빼는 기능을 하는 함수를 정의한다
def add_sub(n1, n2):
    return n1 + n2, n1 - n2
# 반환값이 두 개
```

```python
# 함수 호출 (튜플로 묶여서 출력)
add_sub(5, 2)
```

```python
# 변수를 하나하나 따로 불러줬기 때문에 언패킹돼서 출력
f1, s2 = add_sub(5, 2)
print(f1, s2)
```

##### 가변 매개변수(*)
- 함수 호출 시 몇 개의 인수가 전달될 지 알 수 없다면, 사용자가 직접 매개변수 개수를 정할 수 있도록 선언

```python
def add(*args):
    return sum(args)

# 가변 매개변수를 받아서 모든 숫자를 더한 결과를 반환하는 함수
# sum() 내장함수 이용
```

```python
add(1, 2, 3, 4, 5) # 내가 넣은 값들이 튜플형태로 들어감
```

##### 가변 매개변수(**)
- 내가 넣은 값들을 딕셔너리 형태로 함수 내부를 처리하고 싶을 때

```python
def print_map(**kwargs):
    print(kwargs)
    for k, v in kwargs.items():
        print(k, ':', v)
```

```python
print_map(하나 = 1, 둘 = 2)
```

#### 네임 스페이스(namespace, 이름공간)
- 빌트인 네임스페이스 : 기본 내장 함수/내장 예외들의 이름
    -> 최상위 네임스페이스
- 전역 네임스페이스 : Global, 모듈전체(우리가 선언한 함수에서도)에서 통용될 수 있는 이름
- 지역 네임스페이스 : Local, 함수 내에서만 쓸 수 있음
- 프로그래밍 언어에서 특정한 객체를 이름에 따라 구분할 수 있는 범위

```python
# 함수정의
num = 0 # 전역변수
def temp():
    num = 1 # 지역변수
    print('함수 안 : ', num)
print('함수 밖 : ', num)
temp()
```

```python
# 지역변수를 전역변수로 사용하고 싶을 때(인식)
# global 키워드 사용
num = 0 # 전역변수
print('함수 밖1 : ', num)
def temp():
    global num # num을 전역변수로 사용하겠다!
    num = 1 # 지역변수
    print('함수 안 : ', num)
temp() # 함수를 호출하면 전역변수로 변경이 된다!
print('함수 밖2 : ', num)
```
