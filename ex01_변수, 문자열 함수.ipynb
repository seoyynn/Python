#### 특정문자(열)에 접근하는 방법
- 인덱싱(indexing)
- 슬라이싱(slicing)

```python
# 문자인덱싱
s = 'This subject is python'
# u 접근 : 6번 인덱스 접근
s[6]
# p 접근 : 16번 인덱스, -6번 인덱스
s[-6]
```

```python
# 문자열 슬라이싱
# [첫인덱스:끝인덱스]

print(s[0:4])
print(s[5:12])
print(s[16:22])
```

```python
print(s[13:22])
print(s[-9:]) # 끝 인덱스가 문장 끝이라면 생략 가능
print(s[13:])
```

```python
# 문장 처음부터 끝까지 슬라이싱(첫인덱스, 끝인덱스 생략가능)
# 시작 0이 생략, 끝은  문장길이
s[:]
```

```python
# s[시작인덱스:끝인덱스:증감값]
s[::2]
```

```python
# 거꾸로 출력
s[::-1]
```

#### 문자열 함수

```python
# day 문자열에서 '햇볕 쨍쨍' 인덱스 번호 확인하기
# find 함수 활용
a = day.find('햇볕 쨍쨍')
day[a:a+5]
```

```python
# find 함수를 사용하여 날짜 인덱스 번호 확인하기
# 끝 인덱스를 포함하기 위해 +1
b = day.find('일')
day[:b+1]
```

```python
# index 함수
day.index('일')
```

```python
# find, index 함수 차이 비교 : 문자안에서 없는 문자열 찾을때
day.find('e') # 없을 경우에 -1을 반환, 에러 발생하지 않는다
day.index('e') # 문자를 찾을 수 없다면 에러가 발생한다
```

```python
# Len() : 문자열 개수 반환(길이 반환)
len(day) # 공백, !, ?, . 까지 모두 포함한 결과
```

```python
# replace('찾을 문자열','변경할 문자열') : 특정 문자열 찾아서 변경하는 함수
# day : 햇볕 쨍쨍 -> 비가 주룩
day = day.replace('햇볕 쨍쨍','비가 주룩')
day
```

```python
# strip() : 공백을 제거해 주는 함수
day2 = " 오늘은 월요일입니다. "
day2
```

```python
day2.strip()
```

```python
# lstrip() : 왼쪽 공백 제거
# Rstrip() : 오른쪽 공백 제거
day2.lstrip()
day2.rstrip()
```

```python
# count('찾을 문자열') : '찾을 문자열'이 몇 개 있는지 개수 반환
test = 'today is monday!!!'
test.count('monday')
```

```python
# 대문자 -> 소문자 : Lower
# 소문자 -> 대문자 : Upper
test.upper().count('D')
test.upper()
```

```python
# split() : 문자열 나눠주는 함수
test.split() # 공백을 기준으로 문자열을 나눔
```
