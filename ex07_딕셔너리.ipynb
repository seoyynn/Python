#### 딕셔너리란?
- key와 value를 한쌍으로 갖는 자료형
- 불면한 key와 가변/불면적인 value로 맵핑된 순서가 없는 집합
- 인덱싱, 슬라이싱이 불가능

#### 딕셔너리 선언

```python
# 딕셔너리명 = {key:value, key:value, ...}
a = {} # 빈 딕셔너리
b = {'name' : 'key'} # 이름정보(문자열)를 담은 딕셔너리
c = {1 : 'key', 2 : 23} # 하나의 딕셔너리에는 다양한 자료형을 담을 수 있다
d = {'name' : ['서연', '두식', '조연경바보바보']} # value값으로 리스트(자료구조)를 담을 수 있다
# 명확하게 뭔가를 정의하고 싶을 때, 딕셔너리를 사용할 수 있다
```

```python
print(a)
print(b)
print(c)
print(d)
```

```python
print(type(b))
```

```python
# 위의 정보를 담는 딕셔너리 선언해보기
dic1 = {'name' : 'EY', 'age' : 20, 'phone' : '010-1234-5678'}
```

```python
print(dic1)
```

#### 딕셔너리 값 가져오기(데이터에 접근하기)

```python
dic1
# dic1에서 이름 가져오기
dic1['name']
# 폰 번호 가져오기
dic1['phone']
```

```python
# 데이터에 접근할 때 사용하는 함수 : get(키 값)
dic1.get('name')

# 생일 데이터를 가져오기
dic1.get('babo')
```

- 딕셔너리 접근 방법 비교
    - 딕셔너리명[키명] vs 딕셔너리명.get(키명)

```python
gender = dic1['gender']
gender
```

```python
gender = dic1.get('gender')
print(gender)
```

```python
# 예외처리 try ~ except
try:
    gender = dic1['gender']
    gender
except KeyError:
    print('없는 키 값입니다')
```

##### 딕셔너리 key, value 모두 가져오기

```python
# items()
dic1.items()
```

```python
for i in dic1.items(): # .items() -> key와 value값을 한번에 출력할 수 있음
    print(i)
    
# 튜플형태로 출력됨
```

```python
for key, value in dic1.items():
    print(key, value)
    
# key와 value를 따로 불러오면 아래와 같은 결과가 나옴
# 따로따로 불러주면 튜플로 반환이 되지않음(언패킹)
```

```python
# 총 합을 구하는 함수 : sum()
# 1. python 과목에 접근해서 결과 가져오기
# 2. 1번의 합을 사람 수 만큼 나누기, len()사용

py_avg = 0
py_sum = 0

score_dict = {'이름' : ['장서연', '장두식', '조연경', '박유정', '박서연'],
             'Python' : [90, 100, 85, 90, 80],
             'Java' : [85, 100, 80, 95, 85],
             'html/css' : [75, 70, 90, 80, 90]}

for i in score_dict.items():
    py_sum = sum(score_dict.get('Python'))
    py_avg = py_sum / len(score_dict.get('Python'))
print(py_avg)

```

```python
py_sum
```

```python
# 이름은 빼고 과목만 가져오기 위해 인덱싱, 슬라이싱을 위해 리스트로 형변환
# 선생님 풀이
avg = 0
for i in list(score_dict.keys())[1:]:
    avg = sum(score_dict[i])/len(score_dict[i])
    print(f'{i}평균 :{avg}')
```
